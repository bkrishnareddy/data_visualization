```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Linear Plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## Object oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y)
```

```python
# two figures in the same plot

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.4,0.8])
axes1.set_xlabel('xlabel')
axes1.set_ylabel('ylabel')
axes1.set_title('1st Plot')
axes2 = fig.add_axes([0.6,0.1,0.4,0.8])
axes2.set_xlabel('xlabel')
axes2.set_ylabel('ylabel')
axes2.set_title('2nd Plot')
axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
```

```python
# Subplot--OO approach
fig,axes = plt.subplots(nrows=3,ncols=4)
plt.tight_layout()
```

```python
# Subplot--OO approach
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()

# axes is iteratable
for current_ax in axes:
    current_ax.plot(x,y)
```

### Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,6))
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig("my_picture.png",dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label='X Cubed')
ax.legend() #you can also specifiy the position of the parameter
```

# Plot Appearance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='b', linewidth=.5,alpha=0.5,linestyle='-', marker='+',markersize='15') # can take RGB Hex Code also
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=2,ls='--')

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
